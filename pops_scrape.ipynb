```python
##construct a dictionary of dictionaries

pop_data = {}
count = 0
for pop in list_of_pops:
    
    ##get description of pop
    description=''
    paragraphs = pop.find_all("p")
    number = 2
    paragraph= paragraphs[number].text
    stop=False
    while (not("*Required." in paragraph)) and not stop:
        paragraph = re.sub(r'[^\w\s]','', paragraph)
        #paragraph = re.sub('\s+','', paragraph)
        description += paragraph
        number+=1
        if number<len(paragraphs):
            paragraph = paragraphs[number].text
        else:
            stop = True
    
    disability_access = 'unknown'
    hours = 'unknown'
    amenities_info = 'unknown'
    for information in pop.find_all("aside"):
        if 'General Information' in information.text:
            disability_access = information.text.split('Access For Disabled:')[1]
        
        if 'Hours of Access' in information.text:
            step1 = information.text.split('Hours of Access')[1]
            hours = step1.split("Hours")[0]
            
        if 'Required Amenities' in information.text:
            amenities_info = (information.text).split("equired Amenities")[1]

    ##get image
    imgs = pop.find_all("img")
    if len(imgs) > 0: 
        text = str(imgs[0])
        split1=text.split("data-lazy-src=")[1]
        split2=split1.split(' ')
        img = split2[0].replace('"', '')
    else:
        img = ''
    
    ##get name
    title = pop.find_all("h1")
    if len(title)>1:
        name = title[1].text
    else:
        name = ''
        
    ##get rating
    if len(pop.find_all("strong"))>1 and len(pop.find_all("strong")[1].text)<5:
        rating = pop.find_all("strong")[1].text
    else:
        rating = ''
    
    if name != '':
        pop_data[name]={}
        pop_data[name]['summary']=description
        pop_data[name]['img']=img
        pop_data[name]['rating']=rating
        pop_data[name]['disability_access'] = disability_access
        pop_data[name]['hours'] = hours
        pop_data[name]['amenities_info'] = amenities_info
    
    count+=1

print(len(pop_data))
```

```python
#write to a csv file
import csv
header = list(pop_data[list(pop_data.keys())[0]].keys())
with open('processed_data.csv', 'w', newline="") as f:
    write = csv.writer(f)
    write.writerow(['Name', *header])
    for a, b in pop_data.items():
        write.writerow([a]+[b.get(i, '') for i in header])
```
